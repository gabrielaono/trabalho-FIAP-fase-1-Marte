```python
#import pandas
import pandas as pd

#import file
df = pd.read_csv('telemetria_marte.csv')

for index, linha in df.iterrows():
  TI = linha ["Temp_Interna_C"]
  TE = linha ["Temp_Externa_C"]
  I = str( linha["Integridade_Estrutural"])
  E = linha ["Nivel_Energia_Perc"]
  P = linha ["Pressao_Tanques_PSI"]
  S = linha ["Status_Modulos"]
  t = linha ["Timestamp"]

  print("\nTimestamp",t,":")
  if (18<=TI) and (TI<=27) and (-140<=TE) and (TE<=20) and (I== "1") and (E>85) and (2850<=P) and (P<=3150) and (S== "Nominal"):
    print(" NAVE APTA PARA DECOLAGEM")
  else:
    print(" ABORTAR!\n","Motivo: ")
    if (TI<18) or (TI>27):
        print(" Temperatura interna inadequada \n")
    if (TE<-140) or (TE>20):
        print(" Temperatura externa inadequada \n")
    if (I== "0"):
        print (" Falha na integridade estrutural \n")
    if (E<85):
        print(" Energia insuficiente \n")
    if (P<2850) or (P>3150):
        print(" Pressão inadequada \n")
    if (S=="Alerta"):
        print(" Status de alerta")
```
